### Filtros de imagens

### Pegando imagem da câmera

```python
#camera = cv2.VideoCapture(0)
```

```python
camera = cv2.VideoCapture(0)
while True:
    _, frame = camera.read()
    frame_blur = cv2.medianBlur(frame, 13)
    cv2.imshow('Teste', frame_blur)
    pressedkey = cv2.waitKey(1)
    if pressedkey & 0xFF == ord('q'):
        break
camera.release()
cv2.destroyAllWindows()
```
